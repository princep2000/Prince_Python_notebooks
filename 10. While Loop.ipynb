```python
x=int(input('Enter a number'))#1221
temp=x
s=''
while x!=0:
    r=x%10
    s=s+str(r)
    x=x//10
 
if int(s)==temp:
    print('Pallindrome')
else:
    print('Not Pallindrome')
```

```python
x=int(input('Enter a number'))

if x<2:
    print('Invalid Number')
elif x==2:
    print('Prime')
else:
    for i in range(2,x//2+1):
        if x%i==0:
            print('Not Prime')
            break#Break the whole loop
    
    
    else:
        print('Prime')
```

```python
x=int(input('Enter a number'))

if x<2:
    print('Invalid Number')
elif x==2:
    print('Prime')
else:
    for i in range(2,int(x**0.5)+1):
        if x%i==0:
            print('Not Prime')
            break#Break the whole loop
    
    else:
        print('Prime')
```

```python
for x in range(int(input('Enter Starting Number')),int(input('Enter Stop'))):
    if x<2:
        print('Invalid Number')
    elif x==2:
        print('Prime')
    else:
        for i in range(2,int(x**0.5)+1):
            if x%i==0:
                break#Break the whole loop
        else:
            print(x,end='\t')
```
