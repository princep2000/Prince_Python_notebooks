```python
def print_name(name):
    print(name)
    print_name(name)
```

```python
print_name('sachin')
```

```python
def print_name(name,num=1):
    if num>10:
        return
    else:
        print(name)
        print_name(name,num+1)
```

```python
print_name('sachin')
```

```python
def count(num=1):
    if num>10:
        return 
    else:
        print(num)
        count(num+1)
```

```python
count()
```

```python
def iterate(s,num=0):
    if num==len(s):
        return 
    else:
        print(s[num],end='\t')
        iterate(s,num+1)
```

```python
iterate('sachin')
```

```python
def prime(num,start=2):
    if start==int(num**0.5)+2:
        return True
    elif num%start==0:
        return False
    else:
        return prime(num,start+1)
```

```python
prime(121)
```

```python
prime(127)
```

```python

def factorial(q,i=1,p=1):
    if i==q+1:
        print(p)
    else:
        p*=i
        factorial(q,i+1,p)
factorial(6)
factorial(5)
```

```python
def fibonacci(l=[0,1]):
    if len(l)==10:
        return l
    else:
        l.append(l[-1]+l[-2])
        return fibonacci(l)
    
```

```python
fibonacci()
```

```python
def fibo(num,s=0,e=1):
    if num==0:
        return
    else:
        print(s,end='\t')
        s,e=e,s+e
        fibo(num-1,s,e)
```

```python
fibo(10)
```

```python
def pallindrome(n,r=0):
    if n==0:
        return r
    else:
        m=n%10
        r=r*10+m
        return pallindrome(n//10,r)
```

```python
pallindrome(1221)
```

```python
x=int(input('Enter a number'))

if x==pallindrome(x):
    print('Number is Pallidrome')
else:
    print('Number is not a Pallindrome')
```

```python
def star(r=10):
    if r==0:
        return
    else:
        print('*'*(10-r+1))
        star(r-1)
```
